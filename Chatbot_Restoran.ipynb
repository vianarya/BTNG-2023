```python
import json

# 1. Load Dataset
def load_data():
    with open('menu.json', 'r') as file:
        return json.load(file)

data = load_data()
menu_list = data['menu']

# 2. Fungsi Logika Chatbot
def mcd_chatbot(user_input):
    user_input = user_input.lower()

    # Respon untuk sapaan
    if "halo" in user_input or "hi" in user_input:
        return "Halo! Selamat datang di McDonald's Assistant. Ada yang bisa saya bantu terkait menu dan harga?"

    # Logika pencarian harga berdasarkan nama menu
    found_items = []
    for item in menu_list:
        if item['nama'].lower() in user_input:
            found_items.append(f"{item['nama']} harganya Rp{item['harga']:,}")

    if found_items:
        return "Tentu, berikut informasinya: " + ", ".join(found_items)

    # Logika cek kategori (misal: "tampilkan menu burger")
    if "burger" in user_input:
        burgers = [i['nama'] for i in menu_list if i['kategori'] == 'Burger']
        return "Kami punya menu Burger: " + ", ".join(burgers)

    if "daftar menu" in user_input or "harga" in user_input:
        all_menu = [f"{i['nama']} (Rp{i['harga']:,})" for i in menu_list]
        return "Berikut daftar lengkap menu kami: \n" + "\n".join(all_menu)

    return "Maaf, saya tidak mengerti. Anda bisa tanya harga menu tertentu seperti 'Berapa harga Big Mac?'"

# 3. Simulasi Chat
print("--- McDonald's AI Chatbot Ready ---")
print("(Ketik 'keluar' untuk berhenti)")

while True:
    user_query = input("Anda: ")
    if user_query.lower() == 'keluar':
        print("Bot: Terima kasih sudah mampir di McD!")
        break

    response = mcd_chatbot(user_query)
    print(f"Bot: {response}")
```
